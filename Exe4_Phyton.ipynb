```python
# JOGO DA VIDA DE CONWAY

import random
import copy
import time
matriz = []

for y in range(10):
  linha = []
  for x in range(10):
    linha.append(random.choice(['.','#']))
  matriz.append(linha)

# print(matriz)

era = 1

while True:
  # Imprimir a matriz atual
  print(f'\n   {era}\n')
  era += 1
  for y in range(10):
    print('  ', end='')
    for x in range(10):
      print(matriz[y][x], end=' ')
    print()

  # copia a matriz atual
  matriz_futura = copy.deepcopy(matriz)

  # contagem de vizinhos

  for y in range(10):
    for x in range(10):
      viz_esq = x - 1
      viz_dir = x + 1
      viz_sup = y - 1
      viz_inf = y + 1

      vizinhos = 0

      # contagem de vizinhos superiores
      if y > 0:
        if x > 0:
          if matriz[viz_sup][viz_esq] == '#':
            vizinhos += 1

        if matriz[viz_sup][x] == '#':
          vizinhos += 1

        if x < 9:
          if matriz[viz_sup][viz_dir] == '#':
            vizinhos += 1

      # contagem de vizinhos centrais
      if x > 0:
        if matriz[y][viz_esq] == '#':
          vizinhos += 1
      if x < 9:
        if matriz[y][viz_dir] == '#':
          vizinhos += 1

      # contagem de vizinhos inferiores
      if y < 9:
        if x > 0:
          if matriz[viz_inf][viz_esq] == '#':
            vizinhos += 1

        if matriz[viz_inf][x] == '#':
          vizinhos += 1

        if x < 9:
          if matriz[viz_inf][viz_dir] == '#':
            vizinhos += 1

      if matriz[y][x] == '.' and vizinhos == 3:
        matriz_futura[y][x] = '#'
      if matriz[y][x] == '#' and vizinhos > 3:
        matriz_futura[y][x] = '.'
      if matriz[y][x] == '#' and vizinhos < 2:
        matriz_futura[y][x] = '.'

  matriz = copy.deepcopy(matriz_futura)
  time.sleep(1)
```

```python
'''Faça um programa que o usuário digite um CPF (exemplo: 123.456.789-10) e o sistema
valide se o primeiro digito verificador está correto (no caso do exemplo, o número 1 após
o traço).
Para validar este dígito verificador deve seguir os seguintes passo:
1. Multiplicar cada número do CPF por um valor respectivo:
Ex: CPF 132.465.987-10
1 3 2 4 6 5 9 8 7
X
10 9 8 7 6 5 4 3 2
=
10 27 16 28 36 25 36 24 14
2. Somar os resultados:
10 + 27 + 16 + 28 + 36 + 25 + 36 + 24 + 14 = 216
3. Multiplicar o resultado por 10:
216 x 10 = 2160
4. Obter o resto da divisão do valor por 11:
2160 % 11 = 4
5. Se o resultado anterior for maior que 9 o resultado é 0, se não o resultado anterior é
o resultado (no exemplo: 4)
6. Se o Resultado for igual ao primeiro dígito verificador está correto, senão está
incorreto
CPF do exemplo: 132.465.987-10
1º Digito verificador: 1
Resultado do cálculo: 4
Logo:
CPF inválido
'''

cpf = input('Digite o CPF que deve ser avaliado: ')
soma = 0
mult = 10

for digito in cpf[:9]:
  soma += (int(digito)* int(mult))
  mult -= 1

soma *= 10
soma %= 11

if soma > 9:
  soma = 0


if soma == int(cpf[9]):
  print(f'primeiro digito do {cpf} valido.')
else:
  print(f'{cpf} e invalido.')






```

```python
'''Faça um programa que o usuário digite um CPF (exemplo: 123.456.789-10) e o sistema
valide se os dois dígitos verificadores estão corretos (no caso do exemplo, o número 4 e
número 0 após o traço).
Para validar o primeiro dígito, siga o exemplo executado no exercício 3.
Para validar o segundo dígito verificador deve seguir os seguintes passo:
1. Multiplicar cada número por um valor respectivo + o primeiro dígito verificador:
Ex: CPF 132.465.987-46
1 3 2 4 6 5 9 8 7 4
X
11 10 9 8 7 6 5 4 3 2
=
11 30 18 32 42 30 45 32 21 8
2. Somar os resultados:
11 + 30 + 18 + 32 + 42 + 30 + 45 + 32 + 21 + 8 = 269
3. Multiplicar o resultado por 10:
269 x 10 = 2690
4. Obter o resto da divisão do valor por 11:
2690 % 11 = 6
5. Se o resultado anterior for maior que 9 o resultado é 0, se não o resultado anterior é
o resultado (no exemplo: 6)
6. Se o Resultado for igual ao segundo dígito verificador está correto, senão está
incorreto
CPF do exemplo: 132.465.987-46
2º Digito verificador: 6
Resultado do cálculo: 6
Logo:
CPF válido'''


cpf = input('Digite o CPF que deve ser avaliado: ')
soma = 0
mult = 10

for digito in cpf[:9]:
  soma += (int(digito)* int(mult))
  mult -= 1

soma *= 10
soma %= 11

if soma > 9:
  soma = 0


if soma == int(cpf[9]):
  print(f'primeiro digito do {cpf} valido.')
else:
  print(f'{cpf} e invalido.')
```
